```python
def impute_age(col):
    Age = col[0]
    Pclass= col[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
train["Age"]= train[["Age","Pclass"]].apply(impute_age,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False)
```
